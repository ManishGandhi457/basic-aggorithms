```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('D:\Gandhi\Machine Learning\Linear_Regression\Canada_percapita.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per capita income (US$)',fontsize=20)
plt.scatter(df.year,df['per capita income (US$)'])
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict(1974)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per capita income (US$)',fontsize=20)
plt.scatter(df[['year']],df['per capita income (US$)'])
plt.plot(df[['year']],reg.predict(df[['year']]),color='red')
```

```python
import pickle
```

```python
with open('canada_pickle','wb') as f:
    pickle.dump(reg,f)
```
